```python
import pandas as pd
from dataprofiler import Data, Profiler
import psycopg2
from sqlalchemy import create_engine
import json
import pprint

```

```python
engine = create_engine('postgresql+psycopg2://localhost/hockey')
connection = engine.connect()
career_statistics_df = pd.read_sql_query('SELECT * FROM public.stg_wikipedia__career_statistics', connection)
profile = Profiler(career_statistics_df)
report = profile.report(report_options={"output_format": "pretty"})
printer = pprint.PrettyPrinter(sort_dicts=False, compact=True)
printer.pprint(report)
```

```python
print(json.dumps(report["data_stats"][0], indent=4))
```

```python
import pandas_profiling

profile = career_statistics_df.profile_report(title='Pandas Profiling Report')
   
# save the report as html file
profile.to_file(output_file="pandas_profiling1.html")
   
# save the report as json file
profile.to_file(output_file="pandas_profiling2.json")
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = ["A1", "A2", "B1", "B2", "C1", "C2"],
      color = "blue"
    ),
    link = dict(
      source = [0, 1, 0, 2, 3, 3], # indices correspond to labels, eg A1, A2, A1, B1, ...
      target = [2, 3, 3, 4, 4, 5],
      value = [8, 4, 2, 8, 4, 2]
  ))])

fig.update_layout(title_text="Basic Sankey Diagram", font_size=10)
fig.show()
```
